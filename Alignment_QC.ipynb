# Detect empty droplets -R package

```python
## load R Packages
library(DropletUtils,lib.loc="/oak/stanford/groups/icobos/Akila/Rpackages/")
library(Matrix)
library(ggplot2)
library(dplyr)
library(tidyverse)
library(ggExtra)

list1<-c("input sample file names")

list_count <- length(list1)

print(list_count)

for (i in list_count)
{
    ## load matrix for each sample
    
  m <- readMM(paste0("/scratch/groups/icobos/Akila/Astrocytes/Matrix/",list1[i],".mtx"))
  m <- Matrix::t(m)
  m <- as(m, "dgCMatrix")
    
    ## calculate empty droplets based on barcodes 
    
  tot_counts <- Matrix::colSums(m)  
  bc_rank <- barcodeRanks(m)
  set.seed(200)
  e.out <- emptyDrops(m)
    
    ## test each cell for real/ambient RNA profile
  is.cell <- e.out$FDR <= 0.05
  sum(is.cell, na.rm=TRUE)
  #write e.out
    
  write.csv(bc_rank,paste("/scratch/groups/icobos/Akila/Matrix/rank/",list1[i],".csv"))

  write.csv(e.out,paste("/scratch/groups/icobos/Akila/Matrix/e_out/",list1[i],".csv"))
  mypath <- file.path("/scratch/groups/icobos/Akila/Matrix/logprobability/",paste(list1[i], ".jpg", sep = ""))
    
  #log probaility plot empty droplet testing
  jpeg(file=mypath)
  par(mar=c(5,4,1,1), mfrow=c(1,2), bty="n")
    plot(e.out$Total, -e.out$LogProb, col=ifelse(is.cell, "red", "black"),
    xlab="Total UMI count", ylab="-Log Probability", cex=0.2)
    abline(v = bc_rank@metadata$inflection, col="darkgreen")
    abline(v = bc_rank@metadata$knee, col="dodgerblue")
    legend("bottomright", legend=c("Inflection", "Knee"), bty="n", 
    col=c("darkgreen", "dodgerblue"), lty=1, cex=1.2)
    
    #zoom in low UMI part
    #log probaility plot empty droplet testing
  
    plot(e.out$Total, -e.out$LogProb, col=ifelse(is.cell, "red", "black"),
    xlab="Total UMI count", ylab="-Log Probability", cex=0.2, xlim=c(0,2200), ylim=c(0,2200))
    abline(v = bc_rank@metadata$inflection, col="darkgreen")
    abline(v = bc_rank@metadata$knee, col="dodgerblue")
    dev.off()
    
    #cell_calling plots-empty droplets    
    set.seed(100)
    #set inflection point
    limit <- bc_rank@metadata$inflection  
    all.out <- emptyDrops(m, lower=limit, test.ambient=TRUE)
    
    #write all.out
    write.csv(all.out,paste("/scratch/groups/icobos/Akila/Matrix/all_out/",list1[i],".csv"))
    mypath1 <- file.path("/scratch/groups/icobos/Akila/Matrix/cell_calling/",paste(list1[i], ".jpg", sep = ""))
    jpeg(file=mypath1)
    
    #histogram null hypotheis -low UMI denotes empty droplets!!!!
    hist(all.out$PValue[all.out$Total <= limit & all.out$Total > 0],
    xlab="P-value", main="", col="grey80")
    dev.off()
    
    # cut off write
    both = which(is.cell & e.out$Total >= bc_rank@metadata$inflection)
    filt = m[,both]
    
    #using inflection
    inflect = which(bc_rank$total >= bc_rank@metadata$inflection)
    inflect = m[,inflect]
    
    #using(knee)
    kn = which(bc_rank$total >= bc_rank@metadata$knee)
    kn = m[,kn]
    print(paste(list1[i],"FDR<0.05",nrow(is.cell),ncol(is.cell)))
    #print(paste(list1[i],"FDR<0.05+inflection",nrow(filt)))
    print(paste(list1[i],"Inflection",nrow(inflect),ncol(inflect)))
    print(paste(list1[i],"Knee",nrow(kn),ncol(kn)))
    print(paste(list1[i],"total cell counts",nrow(bc_rank),ncol(bc_rank)))
  }
  
```

# Add empty droplet details in anndata object & Compute Mitochondrial and ribosomal gene frequency

```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata 
from scipy.io import mmread
import scrublet as scr
import doubletdetection
from sklearn.decomposition import TruncatedSVD
import matplotlib
import matplotlib.pyplot as plt
from scipy import sparse, io
from scipy.sparse import csr_matrix


list1=["input file names"]

for i in range(len(list1)):
    
    ##read h5ad file

    adata = anndata.read_h5ad("/scratch/groups/icobos/Akila/counts/metadata/"+str(list1[i])+".h5ad")
   # bdata=anndata.read_h5ad("/scratch/groups/icobos/Akila/unsplice_counts/"+str(list1[i])+".h5ad")
    
    ## read bc_rank,FDR from R emptydroplet
    
    bc_rank=pd.read_csv("/scratch/groups/icobos/Akila/Matrix/rank/"+" "+str(list1[i])+" "+".csv")
    FDR=pd.read_csv("/scratch/groups/icobos/Akila/Matrix/e_out/"+" "+str(list1[i])+".csv")
    
    ## copy information (FDR,rank)inside h5ad obs
    
    adata.obs.index = bc_rank.index
    adata.obs.index = FDR.index
    adata.obs['rank'] = bc_rank['rank']
    adata.obs["droplet_FDR"]=FDR["FDR"]
    adata.obs["total_counts"]=bc_rank["total"]
    
    ## write the metadata along with h5ad
    adata.write("/scratch/groups/icobos/Akila/Astrocytes/Matrix/"+str(list1[i])+".h5ad")
    
       
    
    ### read the h5ad file
    adata = anndata.read_h5ad("/scratch/groups/icobos/Akila/Astrocytes/Matrix/"+str(list1[i])+".h5ad")
    
   
    ## remove cells based on FDR <0.05 (remove emptydroplets)
    adata=adata[adata.obs['droplet_FDR'] < 0.05,:]
    
    ## merge the spliced and unspliced counts and store in nuclear layer
    
    adata.layers["nuclear"]=adata.layers["spliced"]+adata.layers["unspliced"]
    
    ### Calculate QC metrices
    adata.X = csr_matrix(adata.layers['nuclear'])
    
      #mitochondrial genes
    adata.var['mt'] = adata.var_names.str.startswith('MT-') 
    
    # ribosomal genes
    adata.var['ribo'] = adata.var_names.str.startswith(("RPS","RPL"))
    
    # hemoglobin genes.
    adata.var['hb'] = adata.var_names.str.contains(("^HB[^(P)]"))

    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt','ribo','hb'],percent_top=None, log1p=False, inplace=True)
    
    mito_genes = adata.var_names.str.startswith('MT-')
    
    # for each cell compute fraction of counts in mito genes vs. all genes
    # the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
    adata.obs['pct_counts_mt'] = np.sum(
        adata[:, mito_genes].X, axis=1).A1/np.sum(adata.X, axis=1).A1
    
    # add the total counts per cell as observations-annotation to adata
    adata.obs['n_counts'] = adata.X.sum(axis=1).A1

    ##  save the plots
    sc.pl.violin(adata, ['n_genes_by_counts', 'n_counts', 'pct_counts_mt'],jitter=0.4, multi_panel=True)
    plt.savefig("/scratch/groups/icobos/Akila/QC_plots/"+str(list1[i])+".png")
    
    ## count the min_max_gene counts/total UMI counts for each sample(count number of cell after QC)
    ngenes=adata.obs["n_genes_by_counts"]
    med=np.median(ngenes)
    total=adata.obs["total_counts"]
    med1=np.median(total)
    print(list1[i],min(ngenes),max(ngenes),round(sum(ngenes)/len(ngenes)),med,min(total),max(total),round(sum(total)/len(total)),med1,adata.shape)

    #adata.write("/scratch/groups/icobos/Akila/QC_splice/"+str(list1[i])+".h5ad")
    #adata = anndata.read_h5ad("/scratch/groups/icobos/Akila/QC_splice/"+str(list1[i])+".h5ad")
    
```

# Harmony integration and clustering

```python
import scanpy.external as sce
import anndata
import pandas as pd
import scanpy as sc
import numpy as np
from scipy.io import mmread
from sklearn.decomposition import TruncatedSVD
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import csv
adata=anndata.read_h5ad("/oak/stanford/groups/icobos/Akila/Shell/scripts/After_QC_nuclei.h5ad")

#### copy the gene name instead of ENSG id (as index)
adata.var["index1"]=adata.var.index
adata.var["gene_name"]=adata.var["gene_name"].astype('object')
adata.var.loc[adata.var['gene_name'] =="",'gene_name']=adata.var['index1']
adata.var.set_index('gene_name',inplace=True)

#### Filter minimum genes
sc.pp.filter_cells(adata, min_genes=300)
print("After removing min_genes",adata.n_obs, adata.n_vars)


#### scale Nrmalize data
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

### Find HVG
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)

#### PCA 
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata, log=True)

### Harmony integration mention covariates

sce.pp.harmony_integrate(adata, ['Assay','Brain.Region'])

#### check the Elbow plot for PCA
### generate leiden graph based nearsest neighbor

sc.pp.neighbors(adata, use_rep="X_pca_harmony",n_pcs=35)
sc.tl.leiden(adata,resolution = 1.0, key_added = "leiden_1.0")
sc.tl.umap(adata)
sc.pl.umap(adata,color=["Assay", "leiden_1.0"],ncols=2, frameon=False,)
plt.savefig("/scratch/groups/icobos/Akila/integration/harmony/new/min_300/After_Assay_brainregion.png")

### List the major celltype marker genes

marker_genes_dict = {
'Oligodendrocytes':['PLP1','MOG','ST18','MBP',	'OLIG1',	'COL18A1','OPALIN'],
'OPC': ['PDGFRA','VCAN'],
'Astrocytes':['SLC1A2','GFAP','CD44','CHI3L1','GRIA1']
'Microglia':['CSF1R','CX3CR1','TREM2'],
'Endothelial': 	['FLT1',	'ERG'],
'pericytes':	['NOTCH3'],
'VLMC':	['CYP1B1',	'COL1A2'],
'Excitatory':	['SLC17A7','CUX2','FEZF2','THEMIS','RORB','PCP4'],
'Inhibitory':['GAD1','GAD2','ADARB2','LHX6','SST','PVALB','KIT']           
             }


### Annotate the major celltypes based on known markers
sc.pl.dotplot(adata, marker_genes_dict,'leiden_1.0', dendrogram=True,standard_scale="var")
plt.savefig("/scratch/groups/icobos/Akila/integration/harmony/marker_harmony_leiden_1.0_dotplot.png")

### plot the Assay, total_counts
sc.pl.umap(adata,color=["Assay", "leiden_1.0","total_counts"], frameon=False,)
plt.savefig("/scratch/groups/icobos/Akila/integration/harmony/new/min_300/plot_new_leiden1.0_Assay_leiden.png")

#### calculate MArker
adata.uns['log1p']["base"] = None
sc.tl.rank_genes_groups(adata,"leiden_1.0",method='wilcoxon',pts=True,corr_method='benjamini-hochberg')
sc.tl.filter_rank_genes_groups(adata,min_in_group_fraction=0.2,min_fold_change=0.8)
sc.tl.dendrogram(adata,"leiden_1.0")

#### calculate marker and plot them
sc.pl.rank_genes_groups_dotplot(adata,n_genes=25,key='rank_genes_groups_filtered',cmap='bwr',standard_scale='var')

#### Annotate the anndata based on major celltype markers and include them in anndata. obs slot Major_celltypes
#### save the marker genes
result = adata.uns['rank_genes_groups_filtered']
groups = result['names'].dtype.names
ranked_genes_df = pd.DataFrame(
    {group + '_' + key: result[key][group]
     for group in groups for key in ['names', 'pvals_adj', 'logfoldchanges']}
)

ranked_genes_df.to_csv("/scratch/groups/icobos/Akila/markers/markers.csv")    
adata.write("/scratch/groups/icobos/Akila/markers/integrated_harmony_adata.h5ad")
```
